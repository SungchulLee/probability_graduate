```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
c = 3.14 
loc = 0. 
scale = 1.
size = (10000,)
sample = stats.weibull_max(c=c, loc=loc, scale=scale).rvs(size=size)
```

```python
plt.hist(sample, bins=100, density=True);
```
