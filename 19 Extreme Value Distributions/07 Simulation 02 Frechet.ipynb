[Why do the Frechet distributions differ in scipy.stats vs R](https://stackoverflow.com/questions/30783440/why-do-the-frechet-distributions-differ-in-scipy-stats-vs-r)

```python
c = -0.198
size = (10000,)
sample = stats.genextreme(c=c).rvs(size=size)
```

```python
plt.hist(sample, bins=100, density=True);
```
