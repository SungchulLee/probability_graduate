```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10_000
u = np.random.uniform(size=(n,))
plt.hist(u, density=True);
```

```python
(1/3 < u) & (u < 2/3)
```

```python
x = np.zeros_like(u)
x[ (1/3 < u) & (u < 2/3) ] = 1.
x[              u >= 2/3 ] = 2.
plt.hist(x, density=True);
```

```python
x = - np.log(u)
plt.hist(x, density=True, bins=100);
```
