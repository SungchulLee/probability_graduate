# Sample Space $\Omega$
$$
\Omega\quad\text{Sample Space}
$$

# Algebra (or Field) ${\mathcal F}_0$
Let ${\mathcal F}_0$ be a collection of subsets (called events) $A$ of $\Omega$ that has been collected in a certain way. If ${\mathcal F}_0$ satisfies the following three, it is called 
an algebra (or field) over $\Omega$:
$$\begin{array}{lll}
(1)&&\emptyset\in{\mathcal F}_0, \Omega\in{\mathcal F}_0\\
(2)&&\text{$A^c\in{\mathcal F}_0$ for any $A\in{\mathcal F}_0$}\\
(3)&&\text{$A\cup B\in{\mathcal F}_0$ for any $A\in{\mathcal F}_0$ and $B\in{\mathcal F}_0$}\\
\end{array}$$

# Finitely Additive Measure $\mu_0$ 

Let $\mu_0$ be a function defined on an algebra ${\mathcal F}_0$:
$$
\mu_0: {\mathcal F}_0\rightarrow [0,\infty]
$$
If $\mu_0$ satisfies the following, it is called a finitely additive measure over $\Omega$:
For any finite disjoint events $A_1, A_2, \dots, A_N$
$$
\mu_0\left(\bigcup_{n=1}^N A_n\right)=\sum_{n=1}^N \mu_0(A_n)
$$

# $\sigma$-Algebra (or $\sigma$-Field) ${\mathcal F}$
Let ${\mathcal F}$ be a collection of subsets $A$ of $\Omega$ that has been collected in a certain way. If ${\mathcal F}$ satisfies the following three, it is called 
an algebra (or field) over $\Omega$:
$$\begin{array}{lll}
(1)&&\emptyset\in{\mathcal F}, \Omega\in{\mathcal F}\\
(2)&&\text{$A^c\in{\mathcal F}$ for any $A\in{\mathcal F}$}\\
(3)&&\text{$\cup_{i=1}^\infty\in{\mathcal F}$ for any countably infinite $A_i\in{\mathcal F}$}\\
\end{array}$$

# Measure $\mu$ 

Let $\mu$ be a function defined on a $\sigma$-algebra ${\mathcal F}$:
$$
\mu: {\mathcal F}\rightarrow [0,\infty]
$$
If $\mu$ satisfies the following, it is called a measure over $\Omega$:
For any countably infinite disjoint  events $A_1, A_2, \dots$
$$
\mu\left(\bigcup_{n=1}^\infty A_n\right)=\sum_{n=1}^\infty \mu(A_n)
$$

# Hahn–Kolmogorov Theorem (or Carathéodory's Extension Theorem)

Let $\mu_0$ be a finitely additive measure over
an algebra ${\mathcal F}_0$. If $\mu_0$ is countably additive, i.e.,
for any countably disjoint  infinite events $A_1, A_2, \dots$ with $\cup_{n=1}^\infty A_n\in{\mathcal F}_0$
$$
\mu_0\left(\bigcup_{n=1}^\infty A_n\right)=\sum_{n=1}^\infty \mu_0(A_n)
$$
then
there exists a measure $\mu$ defined on the $\sigma$-algebra ${\mathcal F}$
generated by ${\mathcal F}_0$, that is the smallest $\sigma$-algebra containing 
${\mathcal F}_0$, such that for any $A\in{\mathcal F}_0$
$$
\mu(A)=\mu_0(A)
$$
Moreover, if $\mu_0$ is $\sigma$-finite, that is if 
there are countably infinite events $A_1, A_2, \dots$ with 
$$\begin{array}{lll}
(1)&&A_n\in{\mathcal F}_0\\
(2)&&\bigcup_{n=1}^\infty A_n=\Omega\\
(3)&&\mu_0(A_n)<\infty\\
\end{array}$$
then the extension $\mu$ is unique (and also $\sigma$-finite).
